```python
def ks(it, b):
    n = len(it)
    dp = [[0] * (b + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        p = it[i-1][1]  # price
        v = it[i-1][2]  # priority (value)
        for w in range(b + 1):
            if p <= w:
                dp[i][w] = max(dp[i-1][w], dp[i-1][w-p] + v)
            else:
                dp[i][w] = dp[i-1][w]

    # Backtrack to find selected items
    r = []
    w = b
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            r.append(it[i-1])
            w -= it[i-1][1]

    return dp[n][b], r[::-1]

# Main program
it = []
B = int(input("Budget: "))
n = int(input("Items: "))

for _ in range(n):
    name = input("Name: ")
    price = int(input("Price: "))
    priority = int(input("Priority: "))
    category = input("Category (e.g., dairy): ")
    it.append((name, price, priority, category))

dairy_only = input("Only dairy? y/n: ")
if dairy_only == "y":
    it = [i for i in it if i[3] == "dairy"]

v, r = ks(it, B)
print("Max priority:", v)
print("Selected items:")
for item in r:
    print(f"{item[0]} (price: {item[1]}, priority: {item[2]})")
```
