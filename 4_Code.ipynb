```python
import math

def booking_logic(depth, idx, is_max, scores, alpha, beta):
    if depth == 3: return scores[idx]

    best = -math.inf if is_max else math.inf
    for i in range(2):
        val = booking_logic(depth + 1, idx * 2 + i, not is_max, scores, alpha, beta)
        if is_max:
            best, alpha = max(best, val), max(alpha, max(best, val))
        else:
            best, beta = min(best, val), min(beta, min(best, val))
        if beta <= alpha: break
    return best

user_input = input("Enter 8 package scores separated by spaces: ")
package_data = [int(x) for x in user_input.split()]

result = booking_logic(0, 0, True, package_data, -math.inf, math.inf)
print(f"Optimal Choice Score: {result}")
```
