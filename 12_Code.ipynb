```
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Dataset
data = {
    'weight': [10, 50, 30, 12, 45, 35],
    'moisture': [2, 60, 5, 3, 55, 4],
    'density': [0.9, 1.2, 7.8, 0.95, 1.1, 7.5],
    'type': ['Plastic', 'Organic', 'Metal', 'Plastic', 'Organic', 'Metal']
}

df = pd.DataFrame(data)

X = df[['weight', 'moisture', 'density']]
y = df['type']

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# User input
w = float(input("Enter weight: "))
m = float(input("Enter moisture level: "))
d = float(input("Enter density: "))

prediction = model.predict([[w, m, d]])

print("Predicted Waste Type:", prediction[0])
```
