```python
from collections import deque

# Initialize empty graph
graph = {}

# Input: Build graph
n = int(input("Enter number of routers: "))
for i in range(n):
    node = input("Router name: ").strip()
    neighbors = input("Connected routers (space-separated or 'none'): ").strip()
    if neighbors.lower() == 'none':
        graph[node] = []
    else:
        graph[node] = [nb.strip() for nb in neighbors.split()]

start = input("Starting router: ").strip()

# BFS Traversal
visited = set()
queue = deque([start])
visited.add(start)

print(f"BFS Traversal from {start}:")
while queue:
    node = queue.popleft()
    print(node, end=" → ")

    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            visited.add(neighbor)
            queue.append(neighbor)

print("END")
```

```python
graph = {}

n = int(input("Enter number of caves: "))
for i in range(n):
    node = input("Cave name: ").strip()
    neighbors = input("Connected caves (space-separated or 'none'): ").strip()
    if neighbors.lower() == 'none':
        graph[node] = []
    else:
        graph[node] = [nb.strip() for nb in neighbors.split()]

visited = []

def dfs(node):
    if node not in graph:
        print("\nCave not found!")
        return

    visited.append(node)
    print(node, end=" → ")

    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(neighbor)

start = input("Starting cave: ").strip()
print(f"DFS Traversal from {start}:")
dfs(start)
print("END")
```
