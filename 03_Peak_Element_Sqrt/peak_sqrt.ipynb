```python
def analyze_light_full():
    try:

        arr = [int(x) for x in input("Enter light readings (space-separated): ").split()]
        if not arr: return print("No data provided.")


        peak = arr[0]
        for val in arr:
            if abs(val) > abs(peak):
                peak = val


        target = abs(peak)
        low, high, res = 0, target, 0
        while low <= high:
            mid = (low + high) // 2
            if mid * mid <= target:
                res, low = mid, mid + 1
            else:
                high = mid - 1


        print(f"Bulb Found: {peak} units")
        print(f"Darkness Factor (sqrt of magnitude): {res}")
    except ValueError:
        print("Please enter valid integers.")

analyze_light_full()
```
