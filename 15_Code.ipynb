```
import random

# Heuristic function: count conflicts
def heuristic(timetable):
    return len(timetable) - len(set(timetable))  # duplicates = conflicts

# Input
n = int(input("Enter number of exams: "))
slots = int(input("Enter number of timeslots: "))

# Initial random timetable
timetable = [random.randint(1, slots) for _ in range(n)]

# Hill Climbing
while True:
    best = timetable[:]

    for i in range(n):
        for s in range(1, slots + 1):
            neighbour = timetable[:]
            neighbour[i] = s

            if heuristic(neighbour) < heuristic(best):
                best = neighbour

    # Stop if no improvement
    if heuristic(best) >= heuristic(timetable):
        break

    timetable = best

# Output
print("Optimized timetable:", timetable)
print("Conflicts:", heuristic(timetable))
```
