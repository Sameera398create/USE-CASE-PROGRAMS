```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

data = {
"text":[
"Narendra Modi visited Chennai",
"Earthquake hit Japan yesterday",
"Cristiano Ronaldo scored a goal",
"Heavy rains affected Mumbai city"
],
"label":[
"Person",
"Event",
"Person",
"Location"
]
}

df = pd.DataFrame(data)

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["text"])
y = df["label"]

model = LogisticRegression()
model.fit(X,y)

news = input("Enter a news sentence: ")

test = vectorizer.transform([news])
prediction = model.predict(test)

print("Extracted Information:", prediction[0])
```
