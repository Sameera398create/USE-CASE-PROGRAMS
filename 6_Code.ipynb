```python
import math

def minimax_schedule(depth, idx, is_max, scores, a, b):
    if depth == 3: return scores[idx]
    best = -math.inf if is_max else math.inf
    for i in range(2):
        val = minimax_schedule(depth + 1, idx * 2 + i, not is_max, scores, a, b)
        if is_max:
            best = max(best, val)
            a = max(a, best)
        else:
            best = min(best, val)
            b = min(b, best)
        if b <= a: break
    return best

scores = [int(x) for x in input("Enter 8 attendance scores: ").split()]
best_val = minimax_schedule(0, 0, True, scores, -math.inf, math.inf)
print(f"The Best Move (timeslot) ensures {best_val} attendees.")
```
