```python
from collections import deque

def library_system(shelf, target):
    m, n = len(shelf), len(shelf[0])
    flat = sorted([b for row in shelf for b in row])
    sorted_shelf = [flat[i*n : (i+1)*n] for i in range(m)]

    q, visited = deque([(0, 0, 0)]), {(0, 0)}
    while q:
        r, c, d = q.popleft()
        if (r, c) == target: return sorted_shelf, d
        for dr, dc in [(0,1),(1,0),(0,-1),(-1,0)]:
            nr, nc = r+dr, c+dc
            if 0<=nr<m and 0<=nc<n and (nr, nc) not in visited:
                visited.add((nr, nc))
                q.append((nr, nc, d+1))
    return sorted_shelf, -1
rows = int(input("Number of rows: "))
data = [list(map(int, input(f"Row {i+1}: ").split())) for i in range(rows)]
t_r, t_c = map(int, input("Target (row col): ").split())

shelf, dist = library_system(data, (t_r, t_c))
print(f"Organized: {shelf}\nShortest Path: {dist} steps")
```
