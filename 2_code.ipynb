```python
import random

def get_score(S, calories, protein, limit):
    t_cal = sum(c*s for c, s in zip(calories, S))
    t_pro = sum(p*s for p, s in zip(protein, S))
    return t_pro if t_cal <= limit else 0

def meal_prep_hill_climbing(cal, pro, limit, steps):
    n = len(cal)
    S = [random.randint(0,1) for _ in range(n)]

    for _ in range(steps):
        i = random.randint(0, n-1)
        N = list(S)
        N[i] = 1 - N[i]

        if get_score(N, cal, pro, limit) > get_score(S, cal, pro, limit):
            S = N

    return S, get_score(S, cal, pro, limit)


cal = list(map(int, input("Enter calories (space separated): ").split()))
pro = list(map(int, input("Enter protein  (space separated): ").split()))
limit = int(input("Enter calorie limit: "))
steps = int(input("Enter steps: "))

best_s, best_p = meal_prep_hill_climbing(cal, pro, limit, steps)

print("Optimal Menu:", best_s)
print("Total Protein:", best_p)
```
