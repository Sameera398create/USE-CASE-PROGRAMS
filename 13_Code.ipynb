```
import itertools

n = int(input("Enter number of locations: "))
dist = []

print("Enter distance matrix:")
for i in range(n):
    dist.append(list(map(int, input().split())))

cities = list(range(1, n))
min_cost = 9999
best = ()

for p in itertools.permutations(cities):
    cost = dist[0][p[0]]
    for i in range(len(p) - 1):
        cost += dist[p[i]][p[i+1]]
    cost += dist[p[-1]][0]

    if cost < min_cost:
        min_cost = cost
        best = p

print("Best Route:", (0,) + best + (0,))
print("Minimum Distance:", min_cost)
```
